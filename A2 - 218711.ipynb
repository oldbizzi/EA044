(1) qual é produção mensal que atende a demanda exatamente, e com menor custo?


O primeiro passo é definir as **variáveis de decisão** do problema. Em seguida definimos as **restrições** e por fim a **equação-objetivo** (relativa ao custo) que deve ser minimizada.


A função objetivo que trabalharemos é a seguinte:

> $custo\_produzido + custo\_estoque - receita\_demanda$

Sabendo disso precisamos determinar como cada variável é descrita matematicamente

A variável $demanda$ é a mais simples, sendo dada pela demanda mensal multiplicada pelo valor de venda de cada tipo de pneu:

Logo, esta variável é fixa para cada mês.

Para facilitar, vou numerar os meses da seguinte forma:
> junho = 1

> julho = 2

> agosto = 3

Logo, para o mês 1, a demanda é dada por:
> $mes = 1$

>> $receita\_demanda = 4000 * 7 + 1000 * 9$

>> $receita\_demanda = 37000$

> $mes = 2$

>> $receita\_demanda = 8000 * 7 + 5000 * 9$

>> $receita\_demanda = 101000$

> $mes = 3$

>> $receita\_demanda = 3000 * 7 + 5000 * 9$

>> $receita\_demanda = 66000$

Com isto temos agora as três funções objetivo, uma por mês:

> $Custo_1 = custo\_produzido + custo\_estoque - 37000$

>$Custo_2 = custo\_produzido + custo\_estoque - 101000$

>$Custo_3 = custo\_produzido + custo\_estoque - 66000$

Agora é preciso determinar a variável $produzido$:

>$custo\_produzido = custo\_hora * horas$
Neste caso o custo da hora é sempre fixo, cinco reais, no entanto, a quantidade de horas de produção é uma variável, que estará sujeita às restrições a serem determinadas.

Logo, para todos os meses:

> $custo\_produzido = 5*horas$

Podendo ser decomposto em:

> $custo\_produzido = 5*(h_{w,n} + h_{w,f} + h_{l,n} + h_{l,f})$

Representando as horas de cada máquina para cada tipo de material, as horas variam de acordo com o tanto em estoque e o tanto de horas disponíveis no mês.

Podemos impor algumas restrições iniciais:

As horas de cada máquina para cada tipo não pode exceder o limite de horas imposto no enunciado, logo, para o primeiro mês:

> $h_{w,n,1} + h_{w,f,1} <= 700$

> $h_{l,n,1} + h_{l,f,1} <= 1500$

Para o segundo mês:

> $h_{w,n,2} + h_{w,f,2} <= 300$

> $h_{l,n,2} + h_{l,f,2} <= 400$
 
Por fim, para o terceiro mês:

> $h_{w,n,3} + h_{w,f,3} <= 1000$

> $h_{l,n,3} + h_{l,f,3} <= 300$

Por fim, devemos avaliar a variável estoque, tal variável depende do tanto de unidades sobressalentes do mês anterior.

$custo\_estoque = custo\_unidade\_estoque * unidades\_estocadas$

No primeiro mês não teremos unidades estocadas, logo:

> $custo\_estoque_1 = 0$

Esta é a primeira restrição.

> $custo\_estoque_2 = sobressalente_1* 0,1$

>>$sobressalente = produzido - demanda$

>> Separando para cada tipo de material:

>> $ sobressalente_{nylon,2} = (h_{w,n,1} * 0,15^{-1}  + h_{l,n,1}*0,16^{-1})  - 4000 $

>> $sobressalente_{fibra,2} = h_{w,f,1}*0,12^{-1} + h_{l,f,1}*0,14^{-1} - 1000 $

>> $custo\_estoque_2 = (h_{w,n,1} * 0,15^{-1}  + h_{l,n,1}*0,16^{-1} + h_{w,f,1}*0,12^{-1} + h_{l,f,1}*0,14^{-1}  - 5000) * 0,1$ 

> Para o terceiro mês:

>> $ sobressalente_{nylon,3} = (h_{w,n,2} * 0,15^{-1}  + h_{l,n,2}*0,16^{-1})  - 8000 $

>> $sobressalente_{fibra,3} = h_{w,f,2}*0,12^{-1} + h_{l,f,2}*0,14^{-1} - 5000 $

>> $custo\_estoque_3 = (h_{w,n,2} * 0,15^{-1}  + h_{l,n,2}*0,16^{-1} + h_{w,f,2}*0,12^{-1} + h_{l,f,2}*0,14^{-1} - 13000) * 0,1$

Além disto, teremos um custo fixo por unidade, dada por:

> $custo\_nylon = 3,33 * demanda\_nylon$

> $custo\_fibra = 4,13 * demanda\_fibra$

Logo, os custos fixos são dados por:

> $custo\_fixo_1 = 17450$

> $custo\_fixo_2 = 47290$

> $custo\_fixo_3 = 30640$

Este custo fixo deve ser somado ao custo mensal.

Teremos por fim as três funções objetivo relacionando os custos, dadas por:

> Para junho:
>> $Custo_1 = custo\_produzido + custo\_estoque - receita\_demanda$
>> $Custo_1 = 5*(h_{w,n,1} + h_{w,f,1} + h_{l,n,1} + h_{l,f,1}) - 19550$ 

> Para julho:

>> $Custo_2 = 5*(h_{w,n,2} + h_{w,f,2} + h_{l,n,2} + h_{l,f,2}) + (h_{w,n,1} * 0,15^{-1}  + h_{l,n,1}*0,16^{-1} + h_{w,f,1}*0,12^{-1} + h_{l,f,1}*0,14^{-1}  - 5000) * 0,1 - 53710$

> Para agosto:

>> $Custo_3 = 5*(h_{w,n,3} + h_{w,f,3} + h_{l,n,3} + h_{l,f,3}) +  (h_{w,n,2} * 0,15^{-1}  + h_{l,n,2}*0,16^{-1} + h_{w,f,2}*0,12^{-1} + h_{l,f,2}*0,14^{-1} - 13000) * 0,1  - 35360$ 

Estas são as funções que queremos minimizar, de forma a determinar o número de horas de produção, sem excedentes ao final do último mês.

Podemos combinar as três equações resultando em uma única função objetivo a ser minimizada.

> $total\_horas = (h_{w,n,1} + h_{w,f,1} + h_{l,n,1} + h_{l,f,1}) + (h_{w,n,2} + h_{w,f,2} + h_{l,n,2} + h_{l,f,2}) + (h_{w,n,3} + h_{w,f,3} + h_{l,n,3} + h_{l,f,3})$

> $Custo =  5*(total\_horas) + 0,1*[0,15^{-1} * (h_{w,n,1} + h_{w,n,2}) + 0,12^{-1}*(h_{w,f,1} + h_{w,f,2}) + 0,16^{-1}*(h_{l,n,1} + h_{l,n,2}) + 0,14^{-1}*(h_{l,f,1} + h_{l,f,2})]  - 110420$


Agora devemos tratar das restrições:

O estoque inicial é zero como dito, porém, do modo que o sistema foi montado não é preciso impor uma restrição para isto, nem para os estoques dos dois meses seguintes.

Já foram impostas as restrições relativas às horas de produção:

> $h_{w,n,1} + h_{w,f,1} <= 700$

> $h_{l,n,1} + h_{l,f,1} <= 1500$

Para o segundo mês:

> $h_{w,n,2} + h_{w,f,2} <= 300$

> $h_{l,n,2} + h_{l,f,2} <= 400$
 
Por fim, para o terceiro mês:

> $h_{w,n,3} + h_{w,f,3} <= 1000$

> $h_{l,n,3} + h_{l,f,3} <= 300$

Por fim, devemos impor que todas as variáveis são positivas.

Como queremos a produção atendendo exatamente a demanda teremos de acrescentar as restrições referentes à isto.

O número de horas de produção de uma prensa vezes a qunatidade de pneus por hora somado com o estoque deve ser igual a demanda.

>> $h_{w,n} * 0,15^{-1} + h_{l,n} = 15000 $

>>$h_{w,f} * 0,15^{-1} + h_{l,f} = 11000 $

Onde os valores de h correspondem à soma dos três meses de produção para cada tipo de máquina e material.

```

from __future__ import print_function
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('GLOP')

# Criação das variáveis

# Vou criar uma matriz, uma para a máquina Wheel e outra para a máquina Legal
# A primeira coluna representa as horas produzindo Nylon pra cada mês

Wheel = [[1,0],[0,0],[0,0]]
Legal = [[0,0],[0,0],[0,0]]

Wheel[0][0] = solver.NumVar(0,solver.infinity(),'WN1')
Wheel[0][1] = solver.NumVar(0,solver.infinity(),'WF1')
Legal[0][0] = solver.NumVar(0,solver.infinity(),'LN1')
Legal[0][1] = solver.NumVar(0,solver.infinity(),'LF1')

Wheel[1][0] = solver.NumVar(0,solver.infinity(),'WN2')
Wheel[1][1] = solver.NumVar(0,solver.infinity(),'WF2')
Legal[1][0] = solver.NumVar(0,solver.infinity(),'LN2')
Legal[1][1] = solver.NumVar(0,solver.infinity(),'LF2')

Wheel[2][0] = solver.NumVar(0,solver.infinity(),'WN3')
Wheel[2][1] = solver.NumVar(0,solver.infinity(),'WF3')
Legal[2][0] = solver.NumVar(0,solver.infinity(),'LN3')
Legal[2][1] = solver.NumVar(0,solver.infinity(),'LF3')

print(solver.NumVariables())

# restrições
# hwn1 + hwf1 <=700
constraint0 = solver.Constraint(-solver.infinity(), 700)
constraint0.SetCoefficient(Wheel[0][0], 1)
constraint0.SetCoefficient(Wheel[0][1], 1)

# hln1 + hlf1 <= 1500
constraint1 = solver.Constraint(-solver.infinity(), 1500)
constraint1.SetCoefficient(Legal[0][0], 1)
constraint1.SetCoefficient(Legal[0][1], 1)

# hwn2 + hwf2 <=300
constraint2 = solver.Constraint(-solver.infinity(), 300)
constraint2.SetCoefficient(Wheel[1][0], 1)
constraint2.SetCoefficient(Wheel[1][1], 1)

# hln2+ hlf2 <= 400
constraint3 = solver.Constraint(-solver.infinity(), 400)
constraint3.SetCoefficient(Legal[1][0], 1)
constraint3.SetCoefficient(Legal[1][1], 1)


# hwn3 + hwf3 <=1000
constraint4 = solver.Constraint(-solver.infinity(), 1000)
constraint4.SetCoefficient(Wheel[2][0], 1)
constraint4.SetCoefficient(Wheel[2][1], 1)

# hln3 + hlf3 <= 300
constraint5 = solver.Constraint(-solver.infinity(), 300)
constraint5.SetCoefficient(Legal[2][0], 1)
constraint5.SetCoefficient(Legal[2][1], 1)

constraint5 = solver.Constraint(-solver.infinity(), 15000)
for i in range(0,3):
    constraint5.SetCoefficient(Wheel[i][0], 1/0.15)
    constraint5.SetCoefficient(Legal[i][0], 1/0.16)

constraint6 = solver.Constraint(-solver.infinity(),11000 )
for i in range(0,3):
    constraint6.SetCoefficient(Wheel[i][1], 1/0.12)
    constraint6.SetCoefficient(Legal[i][1], 1/0.14)




objective = solver.Objective()
for i in range(0,2):
    objective.SetCoefficient(Wheel[i][0], 5 + 0.1/0.15)
    objective.SetCoefficient(Wheel[i][1], 5 + 0.1/0.12)
    objective.SetCoefficient(Wheel[i][0], 5 + 0.1/0.16)
    objective.SetCoefficient(Wheel[i][1], 5 + 0.1/0.14)
    
objective.SetCoefficient(Wheel[2][0], 5)
objective.SetCoefficient(Wheel[2][1], 5)
objective.SetCoefficient(Wheel[2][0], 5)
objective.SetCoefficient(Wheel[2][1], 5)

objective.SetMaximization()
```
